Apriori Algorithm is a Machine Learning algorithm which is used to gain insight into the structured relationships between different items involved. The most prominent practical application of the algorithm is to recommend products based on the products already present in the user’s cart.

relation of one thing with another

# Support 
support(work)=number of user who done a praticaluar work/total number of users

# confidence
confidence(movie1 --> movie2)=number of people who watch (movie1 and movie2)/total number of people who watch movie1

# lift
lift=confidence/support

# Step
1) set a minimum support and confidence

2) take all the subset in transactions having higher support than minimum support

3) take all the rules of these subsets having hihger confidence than minimum confidence

4)sort the rules by decreasing lift

```python
# optimizating the sell
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset=pd.read_csv("./Market_Basket_Optimisation.csv",header = None)
```

```python
transactions = []
for i in range(0, 7501):
    transactions.append([str(dataset.values[i,j]) for j in range(0, 20)])
```

```python
from apyori import apriori
rules = apriori(transactions, min_support = 0.003, min_confidence = 0.2, min_lift = 3, min_length = 2)
```
